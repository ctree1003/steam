```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('310380.csv')
```

```python
df
```

```python
df['reviews'] = df.reviews.apply(lambda x : re.sub('[^a-zA-Z0-9]', ' ', x)) # 정규표현식을 사용해서 특수문자를 제거
```

```python
for i in range(8770) : 
    df['reviews'][i] = df['reviews'][i].lower() # 소문자로 변경
    df['reviews'][i] = df['reviews'][i].strip() # 좌우 공백 제거
    df['reviews'][i] = " ".join(df['reviews'][i].split()) # 중복된 공백 문자를 제거
```

```python
import nltk
# nltk.download('stopwords') 
# 처음 사용할때 설치 후 다운 받아야 한다
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords

porter = PorterStemmer()
stop = stopwords.words('english')

#공백으로 단어 분리
def tokenizer(text) :
    return text.split()
#PorterStemmer으로 단어 분리
def tokenizer_poter(text):
    return [poter.stem(word) for word in text.split()] 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
import pickle
```

```python
# 주어진 데이터를 단어 사전으로 만들고 각 단어의 빈도수를 계산한 후 벡터화 하는 객체 생성
tfidf = TfidfVectorizer(lowercase=False, tokenizer=tokenizer)
# 문장별 나오는 단어수 세서 수치화, 벡터화해서 학습을 시킨다.
logistic = LogisticRegression(C=10.0, penalty='l2', random_state=0)
pipe = Pipeline([('vect', tfidf), ('clf', logistic)])
```

```python
# 추천은 1 비추천은 0으로 변환
for i in range(8770) : 
    if 'Not Recommended' in df['title'][i]:
        df['title'][i] = '0'
    else:
        df['title'][i] = '1'
```

```python
 df
```

```python
from sklearn.model_selection import train_test_split

# 80%는 학습, 20%는 test
X_train, X_test, y_train, y_test = train_test_split(df['reviews'], df['title'], test_size=0.2, random_state=0)
```

```python
# 학습한다.
pipe.fit(X_train, y_train)

# 학습 정확도 측정
y_pred = pipe.predict(X_test)
print(accuracy_score(y_test, y_pred))

# 학습된 모델을 저장한다.
with open('pipe.dat', 'wb') as fp :
    pickle.dump(pipe, fp)
```

```python
def using_model() :
    with open('pipe.dat', 'rb') as fp:
        pipe = pickle.load(fp)
    import numpy as np
    while True :
        text = input('리뷰를 작성해주세요 :')
        str = [text]
        # 예측 정확도
        r1 = np.max(pipe.predict_proba(str) * 100)
        # 예측 결과
        r2 = pipe.predict(str)[0]
        # 추천은 1 비추천은 0으로 변환
        if r2 == '1' :
            print('긍정적인 리뷰')
        else :
            print('부정적인 리뷰')
           
        print('정확도 : %.3f' % r1)
        break
```

```python
using_model()
```
